    The radiation represented by the smooth curves in the previous graph is called *Bremsstrahlung* (braking radiation) as it is cauced by electron deceleration.  
    The continuous spectrum results from the rapid deceleration of the electrons hitting the target in different ways, some electrons may release all their energy on impact, generating photons of maximum energy i.e. x-rays of minimum wavelength, while others may require multiple collisions to release all their energy, therefore releasing x-rays with a larger wavelength.  
    The **more energy** is released -> **smaller wavelength**
